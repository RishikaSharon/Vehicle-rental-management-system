```python
def customer_menu():
              print("Welcome to VRMS - Vehicle Rent Management System\n")
              cust_id = input("Please enter your Customer ID: ")
              with open('vrms_customers.txt', 'r') as table_cust:
                  found = False
                  for line in table_cust:
                      if line.startswith(cust_id + ','):
                          found = True
                          cust_details = line.strip().split(',')
                          print(f"\nHey {cust_details[1]}, welcome back!\n")
                          break
              if not found:
                  print("Customer not found!")
              else:
                  print("What would you like to do?")
                  print("-" * 30)
                  print("1. View my profile")
                  print("2. View available vehicles")
                  print("3. Book a vehicle")
                  print("4. View my bookings")
                  print("5. Make a payment")
                  print("6. View plans")
                  print("7. I need help")
                  go_on = "yes"
                  while (go_on == "yes"):
                      choice = int(input("\nEnter your choice: "))
                      print()
                      if(choice == 1):
                          view_profile(cust_id)
                      elif(choice == 2):
                          display_vehicles()
                      elif(choice == 3):
                          add_booking(cust_id)
                      elif(choice == 4):
                          display_bookings()
                      elif(choice == 5):
                          add_payment(cust_id)
                      elif(choice == 6):
                          display_plans()
                      elif(choice == 7):
                          support_request(cust_id)
                      else:
                          print("Invalid entry")
                      go_on = input("Continue with the menu? ").lower()
```
